**Importar as dependencias**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels as stats
import scipy.stats as stats
import seaborn as sns
```

**Validar a quantidade de linhas e colunas**

```python
base = pd.read_csv('Cópia-de-Base-003.csv')
base.shape
```

**Comando para exibir os dados**

```python
base.head()
```

**Comando para excluir uma coluna**

```python

base = base.drop(['NomeDepBol'], axis=1)
base = base.drop(['NOMEDEP'], axis=1)
base = base.drop(['DE'], axis=1)
base = base.drop(['CODMUN'], axis=1)
base = base.drop(['TIPOCLASSE'], axis=1)
base = base.drop(['CODESC'], axis=1)
base = base.drop(['CDREDE'], axis=1)
base = base.drop(['TURMA'], axis=1)
base = base.drop(['DEF1'], axis=1)
base = base.drop(['DEF2'], axis=1)
base = base.drop(['DEF3'], axis=1)
base = base.drop(['DEF4'], axis=1)
base = base.drop(['DEF5'], axis=1)
base = base.drop(['Tip_PROVA'], axis=1)
base = base.drop(['Tem_Nec'], axis=1)
base = base.drop(['cad_prova_lG_CN'], axis=1)
base = base.drop(['cad_prova_mat_CH'], axis=1)
base = base.drop(['particip_lg_CN'], axis=1)
base = base.drop(['particip_mat_CH'], axis=1)
base = base.drop(['CLASSE'], axis=1)
```

**Comando para criar a matriz de total de pontos por matéria**

```python
colunas_interesse = [
    'TOTAL_PONTO_LI', 'TOTAL_PONTO_MAT', 'TOTAL_PONTO_LP',
    'TOTAL_PONTO_BIO', 'TOTAL_PONTO_FIS', 'TOTAL_PONTO_QUI',
    'TOTAL_PONTO_HIS', 'TOTAL_PONTO_GEO', 'TOTAL_PONTO_FIL',
    'TOTAL_PONTO_SOC'
]

# Selecionar as colunas de interesse do dataframe original
base_limpo = base[colunas_interesse]

# Converter strings com vírgulas para números, se necessário
base_limpo = base_limpo.apply(lambda x: x.str.replace(',', '.').astype(float) if x.dtypes == 'object' else x)

# Remover colunas com todos os valores NaN
base_limpo = base_limpo.dropna(how='all', axis=1)

# Calcular a matriz de correlação para as colunas de interesse
corr = base_limpo.corr()

# Selecionar apenas a submatriz com as colunas de interesse
corr_focus = corr.loc[colunas_interesse, colunas_interesse]

# Plotar heatmap focado
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_focus,
    cmap='coolwarm',        # Paleta de cores
    annot=True,             # Mostrar valores no heatmap
    fmt='.2f',              # Formato dos números (2 casas decimais)
    annot_kws={"size": 10}, # Tamanho dos números
    cbar_kws={"shrink": 0.8} # Ajuste no tamanho da barra de cores
)

# Configurações de rótulos
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

# Adicionar título
plt.title("Correlação de total de pontos das matérias", fontsize=14)

# Ajuste de layout
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

**Comando para criar a matriz de porcentagem**

```python
colunas_interesse = ['porc_ACERT_LP', 'porc_ACERT_LI', 'porc_ACERT_BIO', 'porc_ACERT_FIS', 'porc_ACERT_QUI', 'porc_ACERT_MAT', 'porc_ACERT_GEO', 'porc_ACERT_HIS', 'porc_ACERT_FIL', 'porc_ACERT_SOC']  # Exemplo


# Remover colunas com todos os valores NaN
base_limpo = base[colunas_interesse].dropna(how='all', axis=1)

# Converter strings para números (tratando vírgulas como separador decimal)
base_limpo = base_limpo.apply(lambda x: x.str.replace(',', '.').astype(float) if x.dtypes == 'object' else x)

# Calcular a matriz de correlação
corr_focus = base_limpo.corr()

# Verificar se a matriz contém valores válidos
if not corr_focus.isnull().all().all():
    # Criar o heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(
        corr_focus,
        cmap='coolwarm',        # Paleta de cores
        annot=True,             # Mostrar valores no heatmap
        fmt='.2f',              # Formato dos números (2 casas decimais)
        annot_kws={"size": 10}, # Tamanho dos números
        cbar_kws={"shrink": 0.8} # Ajuste no tamanho da barra de cores
    )
    # Configurações de rótulos
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(fontsize=10)

    # Título opcional
    plt.title("Heatmap de Correlação - Colunas Selecionadas", fontsize=14)

    # Ajuste de layout
    plt.tight_layout()

    # Mostrar o gráfico
    plt.show()
else:
    # Caso a matriz contenha apenas NaN
    print("A matriz de correlação contém apenas valores NaN. Verifique os dados de entrada.")
```
